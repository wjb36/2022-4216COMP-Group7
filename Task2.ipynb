```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

# Load the data into a Pandas DataFrame
df = pd.read_csv('earthquakeDatasetG7.csv', parse_dates={'Datetime': ['Date', 'Time']}, usecols=['Date', 'Time', 'Magnitude'],nrows=100)
```

```python
# Create the line graph
plt.plot(df['Datetime'], df['Magnitude'])
plt.xlabel('Time')
plt.ylabel('Magnitude')
plt.title('Earthquake Magnitudes over Time')
plt.grid()
plt.show()
```
